```python
def acct_age(x):
    x = x.split(" ")
    year  = int(x[0].replace('yrs',''))
    months = int(x[1].replace('mon',''))/12
    age = year+months
    return round(age,2)
```

```python
def mon(x):
    p1 = x.split(" ")[0][len(x.split(" ")[0])-4::-1]
    if len(p1)==1:
        p=12*int(p1)
    else:
        p=12*int(str(p1)[1]+str(p1)[0])

    s1=x.split(" ")[1][len(x.split(" ")[1])-4::-1]
    if len(s1)==1:
        s=s1
    else:
        s=str(s1)[1]+str(s1)[0]
    months = int(p) + int(s)
    return months
```

```python
df["AVERAGE.ACCT.AGE"]=df["AVERAGE.ACCT.AGE"].apply(lambda x: mon(x))
df["CREDIT.HISTORY.LENGTH"]=df["CREDIT.HISTORY.LENGTH"].apply(lambda x: mon(x))
```
